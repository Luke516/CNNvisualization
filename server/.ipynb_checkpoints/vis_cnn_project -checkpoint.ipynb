```python
# Download an image and read it into a NumPy array, 
def download(url):
    name = os.path.basename(url)
    image_path = tf.keras.utils.get_file(name, origin=url, cache_subdir='images', cache_dir='./')
    img = image.load_img(image_path)
    return image.img_to_array(img)

def urllib_download(url, cache_dir='images/', img_name=''):
    import urllib.request
    opener = urllib.request.build_opener()
    opener.addheaders = [('User-agent', 'Mozilla/5.0')]
    urllib.request.install_opener(opener)
    img_name = os.path.basename(url) if not img_name else img_name
    image_path = urllib.request.urlretrieve(url, cache_dir + os.path.basename(url))
    img = image.load_img(image_path[0])
    return image.img_to_array(img)

# # deprocess (feature space)
def deprocess(img):
    img = img.copy()
    # normalized tensor: center at 0., std at 0.1
    img -= img.mean()
    img /= (img.std() + 1e-5)
    img *= 0.1
    # clip to [0, 1]
    img += 0.5
    img = np.clip(img, 0, 1)
    
    img *= 255.
    return np.clip(img, 0, 255).astype('uint8')

# Display an image
def show(img, title=None):
    plt.figure(figsize=(8, 8))
    plt.grid(False)
    plt.axis('off')
    if len(img.shape) > 3:
        img = img.squeeze(0)
    title = title if title else img.shape
    plt.title(title)
    plt.imshow(img)
```

```python
def load_image(path_to_img, img_shape=(224, 224)):
    img = tf.io.read_file(path_to_img)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    
    max_dim = 512
    shape = tf.cast(tf.shape(img)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim
    new_shape = tf.cast(shape * scale, tf.int32)
    img = tf.image.resize(img, new_shape)
    img = tf.expand_dims(img, 0)
    return img
```

## Activation Maps

```python
def vgg_layers(visualize_layer_names):
    vgg = tf.keras.applications.VGG19(include_top=False, weights='imagenet')
    vgg.trainable = False
    outputs = [vgg.get_layer(name).output for name in visualize_layer_names]
    model = tf.keras.Model([vgg.input], outputs)
    return model
```

```python
visualize_layers = ['block1_conv2', 'block2_conv2', 'block3_conv2', 'block4_conv2', 'block5_conv2']
```

```python
# visualization layers
vgg = vgg_layers(visualize_layers)
```

```python
def visualize_activation_maps(img, model):
    "Expects float img in [0,1]"
    img = tf.cast(img * 255.0, 'uint8').numpy()
    img = tf.keras.applications.vgg19.preprocess_input(img)
    out_img = model(img)
    return out_img
```

```python
out_img = visualize_activation_maps(img, vgg)
```

```python
# display images in row format
def display_in_row(features, figname='', max_images=8, images_per_row=8):
    n_features = features.shape[0]
    n_features = min(n_features, max_images)

    h_size = features.shape[1]
    w_size = features.shape[2]
    n_cols = (n_features // (images_per_row+1)) + 1
    images_per_row = min(n_features, images_per_row)
    
    if features.ndim == 4:
        display_grid = np.ones((h_size * n_cols, images_per_row * w_size, features.shape[-1]), dtype=np.float32)
    else:
        display_grid = np.ones((h_size * n_cols, images_per_row * w_size), dtype=np.float32)
        
    # plot on grid
    for col in range(n_cols):
        for row in range(images_per_row):
            if col * images_per_row + row < n_features:
                channel_image = features[col * images_per_row + row, ...]
                display_grid[col * h_size : (col + 1) * h_size, row * w_size : (row + 1) * w_size] = channel_image
    scale = 1.5 / w_size
    plt.figure(figsize=(scale * display_grid.shape[1], scale * display_grid.shape[0]))
    plt.axis('off')
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='gray')  
    plt.tight_layout()
    plt.show()
```

```python
vis_act = [deprocess(out[0].numpy()[...,:16]).transpose((2, 0, 1)) for out in out_img]
for act in vis_act:
    display_in_row(act)
```

## Filter using gradient ascent
Another easy way to do to insepct the filters learned by convnets is to display the visual pattern that each filter is meant to respond to. This can be done with **gradient ascent in input space**: applying **gradient descent** to the value of the input image of a convnet so as to maximize the response of a specific filter, starting from a blank input image. The resulting input image would be one that the chosen filter is maximally responsive to.

```python
def visualize_filters(model, layer_name, filter_index, size=150, step=40):
    # build a loss function that maximizes the activation 
    # of the nth filter of the layer considered.
    layer_output = model.get_layer(layer_name).output
    vis_model = keras.Model(model.input, layer_output)
    
    # blank input image
    img = tf.Variable(tf.random.normal((1, size, size, 3))) 
    
    epsilon = 1
    for i in range(step):
        with tf.GradientTape() as tape:
            loss = tf.reduce_mean(vis_model(img)[..., filter_index])
        # compute the gradient of the input picutre w.r.t this loss
        grads = tape.gradient(loss, img)
        # Normalize the gradients
        grads /= grads.numpy().std() + 1e-8
        img.assign_add(grads * epsilon)
    return deprocess(img[0].numpy())
```

```python
filter_img = visualize_filters(vgg19, 'block3_conv1', 0, step=40)
show(filter_img)
```

```python
for layer_name in ['block1_conv1', 'block2_conv1', 'block3_conv1', 'block4_conv1']:
    size = 64
    margin = 2
    
    grids = np.zeros((2 * size + 1 * margin, 8 * size + 7 * margin, 3), dtype='uint8')
    for i in range(2):
        for j in range(8):
            filter_img = visualize_filters(vgg19, layer_name, i + (j * 8), size=size)
            # put the result in the square `(i, j)` of the results grid
            horizontal_start = i * size + i * margin
            horizontal_end = horizontal_start + size
            vertical_start = j * size + j * margin
            vertical_end = vertical_start + size
            grids[horizontal_start: horizontal_end, vertical_start: vertical_end, :] = filter_img
            
    plt.figure(figsize=(20, 20))
    plt.imshow(grids)
    plt.show()
```

## Deep Dream using gradient ascent
feature space &rarr; `deprocess` </br>  

input space &rarr; `postprocess_vgg` </br>  

Uses a convolutional neural network to find and enhance patterns in images.  
**Reference**:
- https://www.bonaccorso.eu/wp-content/uploads/2017/07/inceptionism-going-deeper-into-neural.html
- https://github.com/giuseppebonaccorso/keras_deepdream/blob/master/Keras%20Deepdream.ipynb

```python
def postprocess_vgg(img):
    img = img.copy()
    img[..., 0] += 103.939
    img[..., 1] += 116.779
    img[..., 2] += 123.68
    return np.clip(img[..., ::-1], 0, 255).astype('uint8')
```

```python
def calc_loss(layer_activations):
    total_loss = 0
    for act in layer_activations:
        # In gradient ascent, we'll want to maximize this value
        # so our image increasingly "excites" the layer
        loss = tf.math.reduce_mean(tf.square(act))
        
        # normalize by the number of units in the layer
        loss /= np.prod(act.shape)
        total_loss += loss
    return total_loss
```

```python
from IPython.display import clear_output
def deep_dream(img, layer_names, steps=400):
    # build dream model
    layer_output = [vgg19.get_layer(layer).output for layer in layer_names]
    dream_model = keras.Model(vgg19.input, layer_output)
    # preprocess image
    img = keras.applications.vgg19.preprocess_input(img)    
    # make image as variable to be updated.
    input_img = tf.Variable(img)
    
    for step in range(steps):
        with tf.GradientTape() as tape:
            activations = dream_model(input_img)
            loss = calc_loss(activations)
        grads = tape.gradient(loss, input_img)
        grads /= grads.numpy().std() + 1e-8
        input_img.assign_add(grads)
        if (step % 10) == 0:
            clear_output()
            print(f"Step {step}: {loss}")
            show(postprocess_vgg(input_img.numpy()))
            plt.show()
    clear_output()
    show(postprocess_vgg(input_img.numpy()))    
    return input_img.numpy()
```

```python
# input image
img = load_image('images/john-wick-2-dog.jpg') # [0, 1]
img = tf.cast(img * 255., 'uint8').numpy() # [0, 255]
# target layer
layer_names = ['block5_conv1', 'block5_conv2']
```

```python
dream_img = deep_dream(img, layer_names)
```
